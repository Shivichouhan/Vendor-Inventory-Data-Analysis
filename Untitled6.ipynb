```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine

# 1. SETUP
engine = create_engine('sqlite:///Shivani_Analytics_Project.db')
path = '/content'
files = [f for f in os.listdir(path) if f.endswith('.csv')]

# 2. DATA LOADING & CLEANING
print("--- [1/3] LOADING DATA ---")
for file in files:
    df = pd.read_csv(os.path.join(path, file))
    table_name = file.split('.')[0].replace("-checkpoint", "")
    df.to_sql(table_name, con=engine, if_exists='replace', index=False)
print("All tables loaded successfully!")

# 3. ADVANCED BUSINESS ANALYTICS (The 'Real' Work)
print("\n--- [2/3] CALCULATING BUSINESS INSIGHTS ---")

# Analysis A: Total Inventory Value (Hamare paas kitne ka maal hai?)
query_val = "SELECT SUM(onHand * Price) as Total_Value FROM end_inventory"
total_val = pd.read_sql(query_val, con=engine).iloc[0,0]
print(f"Total Inventory Value: ${total_val:,.2f}")

# Analysis B: Top 5 Profitable Brands (Pie Chart Logic)
query_brands = "SELECT Brand, SUM(onHand * Price) as Brand_Value FROM end_inventory GROUP BY Brand ORDER BY Brand_Value DESC LIMIT 5"
df_brands = pd.read_sql(query_brands, con=engine)

# 4. VISUALIZATION DASHBOARD (GitHub par photo dikhane ke liye)
print("\n--- [3/3] GENERATING VISUAL REPORTS ---")

plt.figure(figsize=(15, 6))

# Subplot 1: Bar Chart for Vendors
plt.subplot(1, 2, 1)
query_vend = "SELECT VendorName, SUM(Dollars) as Spend FROM purchases GROUP BY VendorName ORDER BY Spend DESC LIMIT 5"
df_vend = pd.read_sql(query_vend, con=engine)
sns.barplot(x='Spend', y='VendorName', data=df_vend, palette='magma')
plt.title('Top 5 Vendors by Spend')

# Subplot 2: Pie Chart for Brands
plt.subplot(1, 2, 2)
plt.pie(df_brands['Brand_Value'], labels=df_brands['Brand'], autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Brand Value Distribution')

plt.tight_layout()
plt.show()

print("\n--- PROJECT SUMMARY ---")
print("1. Data cleaned and stored in SQL.")
print("2. Vendor spending analyzed.")
print("3. High-value brands identified.")
print("\nSUCCESS: Project is now ready for Portfolio/GitHub!")
```
